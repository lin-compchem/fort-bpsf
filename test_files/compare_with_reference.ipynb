```python
import itertools
combos = itertools.combinations(range(1,10), 2)
combos = list(combos)
fortfi = open('combos', 'r')
fcomb = []
while True:
    line = fortfi.readline()
    if not line: break
    words = line.split()
    myc = tuple([int(i) for i in words[1:]])
    if myc[1] < myc[0]:
        tmp = myc[0]
        myc[0] = myc[1]
        myc[1] = tmp
    fcomb.append(myc)
fcomb.sort()    

print(len(combos))
print(combos)
print(len(fcomb))
for i in range(len(fcomb)):
    print(combos[i], fcomb[i])
    if combos[i] != fcomb[i]:
        raise
```
